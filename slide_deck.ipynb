```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
base_color = sb.color_palette()[0]
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('mdf')
```

## Relationship between ProsperScore and income

>Here it appears that there is a strong correlation between ProsperScore and BorrowerAPR. I am taking this and the preceding scatter graph as confirmation of the theory that Prosper Score sets the APR set for the borrower and hence my next steps will look to anlayse what effects this Prosper rating that the APR is dependent on.

```python
plt.xticks(rotation=35)
sb.barplot(data=df, x='IncomeRange', y='ProsperScore', color=base_color,
           order=['$0', '$1-24,999', '$25,000-49,999',
                  '$50,000-74,999', '$75,000-99,999',
                  '$100,000+']).set_title('Relationship between ProsperScore and income');
```

## Relationship between ProsperScore and Debt

```python
sb.barplot(data=df, x='ProsperScore', y='DebtToIncomeRatio', color=base_color).set_title('Relationship between ProsperScore and Debt');
```

## Relationship between Prosper Score and Income Range -  homeowner status



```python
df_homeowner = df[df['IsBorrowerHomeowner']== True]
df_nothomeowner = df[df['IsBorrowerHomeowner']== False]
plt.xticks(rotation=35)

sb.barplot(data=df_homeowner, x='IncomeRange', y='ProsperScore', color=base_color, order=['$0', '$1-24,999',
                                                                                             '$25,000-49,999',
                                                                                             '$50,000-74,999',
                                                                                             '$75,000-99,999',
                                                                                             '$100,000+']).set_title('Relationship between Prosper Score and Income Range - homeowner');

```

```python
plt.xticks(rotation=35)
sb.barplot(data=df_nothomeowner, x='IncomeRange', y='ProsperScore', color=base_color, order=['$0', '$1-24,999',
                                                                                             '$25,000-49,999',
                                                                                             '$50,000-74,999',
                                                                                             '$75,000-99,999',
                                                                                             '$100,000+']).set_title('Relationship between Prosper Score and Income Range - not a homeowner');
```
